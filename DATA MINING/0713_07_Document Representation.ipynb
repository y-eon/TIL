```
docs = ['오늘 동물원에서 원숭이를 봤어',
        '오늘 동물원에서 코끼리를 봤어 봤어',
        '동물원에서 원숭이에게 바나나를 줬어 바나나를']
```

```
from collections import defaultdict
import numpy as np

word_ls = []
for i in range(len(docs)):
  word_ls.extend(docs[i].split())
word_ls
```

```
word_dict = {}
index = 0
for word in word_ls:
  if word not in word_dict.keys():
    word_dict[word] = index
    index += 1
word_dict
```

```
from collections import defaultdict

word2index = defaultdict(lambda : len(word2index))
for word in word_ls:
  word2index[word]
```

```
word2index
```

```
unique_word_size = len(word2index)
one_hot_vectors = np.zeros((len(word_ls),unique_word_size)) #2차원
one_hot_vectors
```

```
for i, word in enumerate(word_ls):
  index = word2index[word] #우리가 만든 dict형태에서 찾는다
  one_hot_vectors[i, index] = 1 # onehot에서 1인 것을 찾아넣어주는 것

one_hot_vectors
```

```
np.sum(one_hot_vectors,axis=0)
```

### 강사님 버전

```
docs = ['오늘 동물원에서 원숭이를 봤어',
        '오늘 동물원에서 코끼리를 봤어 봤어',
        '동물원에서 원숭이에게 바나나를 줬어 바나나를']
```

```
doc_ls = [doc.split() for doc in docs]
doc_ls
```

```
doc_ls = []
for doc in docs:
  doc_ls.append(doc.split())
doc_ls
```

```
from collections import defaultdict

word2id = defaultdict(lambda : len(word2id))

# [word2id[token] for doc in doc_ls for token in doc] -> 아래랑 같은 것
for doc in doc_ls:
    for token in doc:
        word2id[token]
word2id
```

```
import numpy as np

BoW_ls = []
for i, doc in enumerate(doc_ls):
  bow = np.zeros(len(word2id),dtype=int)
  for token in doc:
    bow[word2id[token]] += 1
  BoW_ls.append(bow)
```

```
BoW_ls
```

```
from IPython.core import display as ICD
import pandas as pd

sorted_vocab = sorted((value, key) for key, value in word2id.items())
vocab = [v[1] for v in sorted_vocab]
for i in range(len(docs)):
  print("문서 {} : {}".format(i, docs))
  ICD.display([pd.DataFrame([BoW_ls[i]],columns=vocab)])
  print("\n\n")
```
